```python
from Data import Data
import pandas as pd
import requests
import json
from pathlib import Path
import pandasdmx as sdmx
```

```python
import sys
ausbs = sdmx.Request('ABS_XML')
data_msg = ausbs.data('RES_DWELL_ST')
data=data_msg.data[0]
```

```python
set(series_key.REGION for series_key in data.series.keys())
```

```python
# Setup regions to pull data, MEASURE= '1' means 'Value of dwelling stock, owned by All Sectors'
AUS = [s for sk, s in data.series.items() if sk.MEASURE =='1']
```

```python
# Convert sdmx to dataframe
cur_df = pd.concat(sdmx.to_pandas(AUS)).unstack()
cur_df
```

```python
# Data cleaning and transpose
cur_df = cur_df.loc['1']
cur_df = cur_df.transpose()
cur_df = cur_df.reset_index()
cur_df.head()
```

```python
cur_df = cur_df.droplevel(1, axis=1)
cur_df
```

```python
# Setup and input data into new dataframe property_value_df 
property_value_df = cur_df
```

```python
#setup index
property_value_df = property_value_df.apply(lambda x: x.replace({'Q1':'Mar', 'Q2':'Jun', 'Q3':'Sep', 'Q4':'Dec'}, regex=True))
property_value_df = property_value_df.rename(columns={'1':'Sydney', '2':'Melbourne','3':'Brisbane', '4':'Adelaide', '5':'Perth', '6':'Hobart','7':'Darwin','8':'Canberra'})
property_value_df['TIME_PERIOD'] = pd.to_datetime(property_value_df['TIME_PERIOD'])
property_value_df.set_index('TIME_PERIOD',inplace=True)
property_value_df.head()
```

```python
# Read csv RBA_Cash Rate_monthly average.csv
ir_df = pd.read_csv(Path('Resources/RBA_Cash Rate_monthly average.csv'))
ir_df.head(15)
```

```python
#Clean the unneeded columns
ir_df = ir_df.iloc[10:,0:2].dropna()
ir_df.columns = ['TIME_PERIOD','Rate']

display(ir_df.head())
display(ir_df.tail())
```

```python
#Change index to match the same format as the property prices dataframe
ir_df['TIME_PERIOD'] = pd.to_datetime(ir_df['TIME_PERIOD'])
ir_df.set_index('TIME_PERIOD',inplace=True)
ir_df.head()
```

```python
# Concatenate the Property Price and Interest Rates DataFrames

all_data_api = property_value_df.merge(ir_df, how='inner', on = 'TIME_PERIOD')
display(all_data_api.head())
display(all_data_api.tail())
```
