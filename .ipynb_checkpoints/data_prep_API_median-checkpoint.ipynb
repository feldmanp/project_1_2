```python
from Data import Data
import pandas as pd
import requests
import json
from pathlib import Path
import pandasdmx as sdmx
%store -r
```

```python
import sys
ausbs = sdmx.Request('ABS_XML')
data_msg = ausbs.data('RES_DWELL')
data=data_msg.data[0]
```

```python
set(series_key.REGION for series_key in data.series.keys())
```

```python
# Setup regions to pull data, MEASURE= '3' means 'Median Price of Established Dwelling Transfers'
AUS = [s for sk, s in data.series.items() if sk.MEASURE =='3']
```

```python
# Convert sdmx to dataframe
median_est_df = pd.concat(sdmx.to_pandas(AUS)).unstack()
median_est_df
```

```python
# Data cleaning and transpose
median_est_df = median_est_df.loc['3']
median_est_df = median_est_df.transpose()
median_est_df = median_est_df.reset_index()
median_est_df.head()
```

```python
# Drop FREQ level
median_est_df = median_est_df.droplevel(1, axis=1)
median_est_df
```

```python
#setup index
median_est_df = median_est_df.apply(lambda x: x.replace({'Q1':'Mar', 'Q2':'Jun', 'Q3':'Sep', 'Q4':'Dec'}, regex=True))
median_est_df['TIME_PERIOD'] = pd.to_datetime(median_est_df['TIME_PERIOD'])
median_est_df.set_index('TIME_PERIOD',inplace=True)
median_est_df.head()
```

```python
median_est_df = median_est_df.dropna()
median_est_df.head()
```

```python
# Read csv RBA_Cash Rate_monthly average.csv
ir_df = pd.read_csv(Path('Resources/RBA_Cash Rate_monthly average.csv'))
ir_df.head(15)
```

```python
#Clean the unneeded columns
ir_df = ir_df.iloc[10:,0:2].dropna()
ir_df.columns = ['TIME_PERIOD','Rate']

display(ir_df.head())
display(ir_df.tail())
```

```python
#Change index to match the same format as the property prices dataframe
ir_df['TIME_PERIOD'] = pd.to_datetime(ir_df['TIME_PERIOD'])
ir_df.set_index('TIME_PERIOD',inplace=True)
ir_df.head()
```

```python
# Concatenate the Property Price and Interest Rates DataFrames

all_data_est = median_est_df.merge(ir_df, how='inner', on = 'TIME_PERIOD')
display(all_data_est.head())
display(all_data_est.tail())
```
