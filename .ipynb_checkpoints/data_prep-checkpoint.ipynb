```python
from Data import Data
import pandas as pd
import requests
import json
from pathlib import Path
```

```python
property_value_df = pd.read_csv(Path("Resources/res_dwell_st.csv"))
property_value_df.head()
```

```python
#Property values CSV cleanup
property_value_df=property_value_df[property_value_df['UNIT_MEASURE']=='AUD']
property_value_df=property_value_df[property_value_df['MEASURE']==1]
property_value_df = property_value_df[['REGION','TIME_PERIOD','OBS_VALUE']]
```

```python
#setup index
property_value_df = property_value_df.apply(lambda x: x.replace({'Q1':'Mar', 'Q2':'Jun', 'Q3':'Sep', 'Q4':'Dec'}, regex=True))
property_value_df['TIME_PERIOD'] = pd.to_datetime(property_value_df['TIME_PERIOD'])
property_value_df.set_index('TIME_PERIOD',inplace=True)
property_value_df.head()
```

```python
# Move every column to be a region
property_value_df = property_value_df.replace({'REGION': {'1':'Sydney', '2':'Melbourne','3':'Brisbane', '4':'Adelaide', '5':'Perth', '6':'Hobart','7':'Darwin','8':'Canberra'}})
propery_values_pivot = property_value_df.pivot(columns='REGION')['OBS_VALUE']
propery_values_pivot.head()
```

```python
ir_df = pd.read_csv(Path('Resources/RBA_Cash Rate_monthly average.csv'))
ir_df.head(15)
```

```python
#Clean the unneeded columns
ir_df = ir_df.iloc[10:,0:2].dropna()
ir_df.columns = ['TIME_PERIOD','Rate']

display(ir_df.head())
display(ir_df.tail())
```

```python
#Change index to match the same format as the property prices dataframe
ir_df['TIME_PERIOD'] = pd.to_datetime(ir_df['TIME_PERIOD'])
ir_df.set_index('TIME_PERIOD',inplace=True)
ir_df.head()
```

```python
# Concatenate the Property Price and Interest Rates DataFrames

all_data = propery_values_pivot.merge(ir_df, how='inner', on = 'TIME_PERIOD')
display(all_data.head())
display(all_data.tail())
```
